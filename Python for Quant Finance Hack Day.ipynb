# Technical Analysis with Quandl

```python
def PlotStock(ticker, MA1='Null', MA2='Null'):
    url = 'GOOG/NASDAQ_'+str(ticker.upper())
    df = pd.DataFrame(Quandl.get(url)) #Don't forget to add your authtoken
    fig = plt.figure(figsize= (12,12))
    plt.subplot(2,1,1)
    df['Close'].plot()
    plt.legend()
    if MA1 == 'Null':
        pass
    else:
        df['MA1'] = pd.rolling_mean(df['Close'],MA1)
        df['MA1'].plot()
    if MA2 == 'Null':
        pass
    else:
        df['MA2'] = pd.rolling_mean(df['Close'], MA2)
        df['MA2'].plot()
    plt.subplot(2,1,2)
    df['Volume'].plot()
    plt.legend()
```

```python
PlotStock('tsla', 500,200)
```

```python
def RelativePerfm(ticker1, ticker2, startdate, enddate):
    url1 = 'GOOG/NASDAQ_'+str(ticker1.upper())
    url2 = 'GOOG/NASDAQ_'+str(ticker2.upper())
    df1 = pd.DataFrame(Quandl.get(url1,
                                  trim_start= startdate,
                                  trim_end= enddate))
    df2 = pd.DataFrame(Quandl.get(url2
                                  , trim_start= startdate
                                  , trim_end= enddate))
    fig = plt.figure(figsize= (12,8))
    df1['relclose1']= df1['Close']/df1['Close'][0]
    df2['relclose2']= df2['Close']/df2['Close'][0]
    df1['relclose1'].plot()
    df2['relclose2'].plot()
    plt.legend()
    plt.title('Relative Performance Comparison')
    
    
```

```python
RelativePerfm('goog', 'aapl', '2010-01-01', '2012-12-12')
```
